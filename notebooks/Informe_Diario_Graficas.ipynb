# INFORME DIARIO

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn-whitegrid')
import seaborn as sns
```

```python
datos=pd.read_csv(r'./Backtesting/datos3.csv')
datos.columns
```

# Grafica de relacion victorias frente a derrotas por accion

```python
dinero_perdido=datos['Dinero Perdido']*-1
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.bar(datos['Stock'],datos['Dinero Ganado'],color='green',width=0.5)
axes.bar(datos['Stock'],dinero_perdido,color='red',width=0.6)
axes.set_title('Relacion Victorias y Derrotas por Accion',fontsize=18,fontweight='bold')
axes.set_xlabel('Stock')
axes.set_ylabel('Dinero ($)')
axes.grid(color='grey', linestyle='dotted', linewidth=0.75)
plt.tight_layout()
plt.show()
```

# Graficos de Densidad por Dinero ganado y perdido

```python
fig, (ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
sns.kdeplot(datos['Dinero Ganado'],ax=ax1,shade=True,color='green')
sns.kdeplot(datos['Dinero Perdido'],ax=ax2,shade=True,color='red')
ax1.set_title('Densidad de Dinero Ganado',fontsize=18,fontweight='bold')
ax2.set_title('Densidad de Dinero Perdido',fontsize=18,fontweight='bold')
plt.tight_layout()
plt.show()
```

# Frecuencias de Dinero Ganado

```python
fig, (ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
sns.histplot(datos['Dinero Ganado'],ax=ax1,stat='frequency',color='green')
sns.histplot(datos['Dinero Perdido'],ax=ax2,stat='frequency',color='red')
ax1.set_title('Histograma Dinero Ganado',fontsize=18,fontweight='bold')
ax2.set_title('Histograma Dinero Perdido',fontsize=18,fontweight='bold')
```

# CONTEO DE FRECUENCIAS POR ACCION

```python
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
sns.countplot(data=datos, x='Stock',ax=axes,palette='bright')
axes.set_title('Cuenta de Cuantos Trades se hizo por Accion',fontsize=18,fontweight='bold')
axes.set_xlabel('Accion')
axes.set_ylabel('Cuantos Trades')
```

# Ganada / Perdida

```python
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
sns.countplot(data=datos, x='Ganada o Perdida',ax=axes,palette='bright')
axes.set_title('Trades Ganados frente a los Perdidos',fontsize=18,fontweight='bold')
axes.set_xlabel('Ganada o Perdida')
axes.set_ylabel('Cuantos Trades')
plt.tight_layout()
plt.show()
```

# Relacion entre el dinero ganado y el spread del broker

```python
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
sns.kdeplot(data=datos, x='Dinero Ganado',y='Spread Etoro ($)',ax=axes,color='green',palette='coolwarm')
axes.set_title('Relacion Dinero Ganado y Spread Etoro',fontsize=18,fontweight='bold')
axes.set_xlabel('Dinero Ganado')
axes.set_ylabel('Spread Etoro ($)')
plt.tight_layout()
plt.show()
```
